```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataFIFA2002 = pd.read_csv("data\FIFA - 2002.csv")
dataFIFA2002.head()
```

```python
dataFIFA2022 = pd.read_csv("data\FIFA - 2022.csv")
dataFIFA2022.head()
```

### **A**

```python
data2002_no_nulls = dataFIFA2002.dropna()
data2022_no_nulls = dataFIFA2022.dropna()
```

### **B**

```python
DataFIFA_FULL = pd.concat([data2002_no_nulls,data2022_no_nulls])

DataFIFA_FULL
```

### **C**

```python
data_3_victorias = DataFIFA_FULL[DataFIFA_FULL["Win"]==3]
data_3_victorias
```

### **D**

```python
Goleadores = DataFIFA_FULL.sort_values(by="Goals For", ascending=False)
Goleadores.set_index("Team", inplace=True)
Goleadores.head(20)["Goals For"].plot(x=None, kind="bar")
```

### **E**

```python
data2002_no_nulls.to_csv("data\FIFA_sin_nulos_2002.csv")

data2022_no_nulls.to_csv("data\FIFA_sin_nulos_2022.csv")
```
